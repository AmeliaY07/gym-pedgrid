```python
import matplotlib.pyplot as plt
import pandas as pd
import pickle5 as pickle
import seaborn as sns
```

```python
df = pd.read_csv("DML_p=1.0.csv")
```

```python
df
```

```python
k = df.loc[df["Direction Split"] == 0.5]
```

```python
plt.plot(k.Density, k["Average Speed"])
```

```python
fig, ax = plt.subplots(figsize = (10,10))
for split in [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]:
    k = df.loc[df["Direction Split"] == split]
    ax.plot(k.Density, k["Average Speed"])
    
ax.plot()
```

```python
sns.relplot(
    data=df, kind="line",
    x="Density", y="Average Speed",
    hue="Direction Split", 
    facet_kws=dict(sharex=False),
)
```
